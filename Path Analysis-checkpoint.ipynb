```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from sklearn import preprocessing
import numpy as np
from sklearn import cross_validation
from sklearn import metrics
```

```python
data = pd.read_csv("..//data//analytical_data_cleaned.csv",encoding = "ISO-8859-1")
```

```python
col_map = pd.read_csv("..//data//question_theme.csv",encoding = "ISO-8859-1")

data.to_csv("..//data//analytical_data_cleaned_new.csv")
col_map.to_csv("..//data//question_theme_new.csv")
```

```python
for con in list(data_orignal['Country'].unique()):
    data = data_orignal.copy()
    data=data[data['Country']==con]

    def yes_no_label(x):
        if (x=="No"):
            return 0
        elif (x=="Yes"):
            return 1
        else:
            return 2

    for col in col_map[col_map['Theme']=="Chronic condition"]['Column'].tolist():
        data[col]=data[col].apply(yes_no_label)

    data=data.drop(['DOB','Age_bucket','Monthly_household_income','Flu_jab_last_6_months','Flu jab frequency',
                    'Chronic_conditions_None'],axis=1)

    labeled_data=data.copy()
    for col in list(labeled_data.columns):
        label_inter = pd.DataFrame(labeled_data[col].copy())
        label_inter['Column_name']= [col]*len(label_inter)
        label_inter.columns=['Orignal_val','Column_name']
        label_inter = label_inter[['Column_name','Orignal_val']]
        try:
            LE_Var = preprocessing.LabelEncoder()
            LE_Var.fit(labeled_data[col])
            labeled_data[col]=LE_Var.transform(labeled_data[col])
            label_inter['New_val']=labeled_data[col]
        except:
            labeled_data[col]=labeled_data[col].astype(str)
            LE_Var = preprocessing.LabelEncoder()
            LE_Var.fit(labeled_data[col])
            labeled_data[col]=LE_Var.transform(labeled_data[col]) 
            label_inter['New_val']=labeled_data[col]
        try:
            Column_val_map = pd.concat([Column_val_map,label_inter])
        except:
            Column_val_map =label_inter.copy()

    Column_val_map.to_csv("..//output//Column_Value_Mapping.csv")

    final_data=labeled_data.copy()
    final_data['Adherence_Score']=np.where(final_data['Adherence_Score']>=2,1,0)

    data_x =  final_data.drop(['Adherence_Score'],axis=1)
    data_y = final_data['Adherence_Score']

    X_train, X_test, y_train, y_test = cross_validation.train_test_split(data_x, data_y, test_size=0.2, random_state=42)

    from sklearn import svm
    clf = svm.SVC(kernel="linear")


    clf.fit(X_train, y_train) 
    y_pred = clf.predict(X_test)

    print("Overall Accuracy = %s" %metrics.accuracy_score(y_pred,y_test))
    print("Classification Matrix")
    print(metrics.classification_report(y_pred,y_test))

    imp= pd.DataFrame(list(data_x.columns))

    imp['coefficient']=clf.coef_[0]

    svm_weights_selected = (clf.coef_ ** 2).sum(axis=0)
    svm_weights_selected /= svm_weights_selected.max()


    imp['importance']=svm_weights_selected

    imp.columns=['Column_name','Coefficient','Importance']

    imp['Filter']="Overall"
    imp['Value']="Overall"


    float(imp[imp['Column_name'].isin(['Confident_I_can_get_a_flu_jab_if_I_want_one'])]['Coefficient'])

    list(labeled_data.columns)

    data_4A=labeled_data[['Confident_I_can_get_a_flu_jab_if_I_want_one','Easy_for_me_to_get_to_a_place_where_I_can_get_the_flu_jab',
                          'I_can_make_time_to_get_the_flu_jab','Can_afford_flu_shot']]

    data_4A=data_4A.fillna(0)

    data_4A['combined_index']=0.0000
    for col in list(data_4A.columns):
        print(col)
        data_4A['combined_index']=data_4A['combined_index']+data_4A[col].astype(float)

    theme_data = data_4A[['combined_index']]

    theme_data.columns=['Availablity']

    Anticipated_Regret=labeled_data[['Worried_about_passing_flu_to_other_people_if_no_flu_jab','Would_regret_not_getting_the_jab']]

    Anticipated_Regret['combined_index']=0.0000
    for col in list(Anticipated_Regret.columns):
        print(col)
        Anticipated_Regret['combined_index']=Anticipated_Regret['combined_index']+Anticipated_Regret[col].astype(float)

    theme_data['Anticipated_Regret'] = Anticipated_Regret[['combined_index']]

    attitude = labeled_data[['In_uncertain_times_I_usually_expect_the_best',
                             'I_am_confident_I_can_achieve_anything_I_set_my_mind_on',
                             'If_something_can_go_wrong_for_me,_it_will','Willingness_to_take_risks']]

    attitude['combined_index']=0.0000
    for col in list(attitude.columns):
        print(col)
        attitude['combined_index']=attitude['combined_index']+attitude[col].astype(float)

    theme_data['attitude'] = attitude[['combined_index']]

    barriers = labeled_data[['Having_the_flu_jab_every_year_may_be_dangerous_for_me',
                            'Felt_strong_flu-like_symptoms_after_getting_a_flu_jab',
                            'I_am_scared_of_getting_a_flu_jab',
                            'The_flu_jab_could_give_me_the_flu',
                            'Contents_of_the_flu_jab_may_be_dangerous_for_me',
                            'Allergic_to_the_flu_jab',
                            'Flu_jab_is_painful',
                            'Worried about the contents in flu jab',
                            'Had the flu this winter so dont need flu jab',
                            'Worried about the side effects of flu jab']]

    barriers['combined_index']=0.0000
    for col in list(barriers.columns):
        print(col)
        barriers['combined_index']=barriers['combined_index']+barriers[col].astype(float)

    theme_data['barriers'] = barriers[['combined_index']]

    cues_to_action = labeled_data[['Remember_to_have_the_flu_jab_on_my_own',
                                  'Remember_if_I_am_feeling_vulnerable_to_illness',
                                  'Remember_when_I_see_it_on_the_news_(such_as_on_TV,_radio_or_in_newspapers)',
                                  'Remember_when_I_see_adverts_in_pharmacies_or_my_doctor\x92s_office',
                                  'My_doctor_reminds_me_or_sends_a_reminder',
                                  'Employer_reminds_me_or_sends_me_a_reminder',
                                  'Informed_about_whether_I_should_or_could_get_the_flu_jab']]

    cues_to_action['combined_index']=0.0000
    for col in list(cues_to_action.columns):
        print(col)
        cues_to_action['combined_index']=cues_to_action['combined_index']+cues_to_action[col].astype(float)

    theme_data['cues_to_action'] = cues_to_action[['combined_index']]

    doctor_rel = labeled_data[['Participation_with_doctor',
                              'My_doctor_thinks_I_should_get_a_flu_jab']]

    doctor_rel['combined_index']=0.0000
    for col in list(doctor_rel.columns):
        print(col)
        doctor_rel['combined_index']=doctor_rel['combined_index']+doctor_rel[col].astype(float)

    theme_data['doctor_rel'] = doctor_rel[['combined_index']]

    trust_info_chan = labeled_data[['Trust_the_information_provided_by_Health_Authorities',
                                   'Trust_the_information_provided_by_news_reports_on_TV_&_radio_or_newspapers_about_the_flu_jab',
                                   'Trust_information_provided_by_general_websites,_YouTube,_blogs,_Facebook_or_Twitter_about_the_flu_jab',
                                   'Trust_the_information_provided_by_my_doctor_about_the_flu_jab']]

    trust_info_chan['combined_index']=0.0000
    for col in list(trust_info_chan.columns):
        print(col)
        trust_info_chan['combined_index']=trust_info_chan['combined_index']+trust_info_chan[col].astype(float)

    theme_data['trust_info_chan'] = trust_info_chan[['combined_index']]

    knowledge = labeled_data[['Seek_information_about_health_issues_that_worry_me',
                             'Knowledge_of_flu']]

    knowledge['combined_index']=0.0000
    for col in list(knowledge.columns):
        print(col)
        knowledge['combined_index']=knowledge['combined_index']+knowledge[col].astype(float)

    theme_data['knowledge'] = knowledge[['combined_index']]

    risk_perc = labeled_data[['Flu_could_make_me_severely_ill',
                              'If_I_got_the_flu,_I_would_feel_sicker_than_other_people_my_age',
                              'If_I_got_the_flu,_I_worry_that_I_would_miss_out_on_important_activities_or_events',
                              'I_believe_that_if_I_got_the_flu_I_would_have_to_stay_in_bed_for\x85',
                              'With_no_flu_jab,_I_would_feel_very_vulnerable_to_the_flu',
                              'Confident_I_can_avoid_getting_the_flu,_even_without_the_flu_jab']]

    risk_perc['combined_index']=0.0000
    for col in list(risk_perc.columns):
        print(col)
        risk_perc['combined_index']=risk_perc['combined_index']+risk_perc[col].astype(float)

    theme_data['risk_perc'] = risk_perc[['combined_index']]

    vacc_trust = labeled_data[['Favouring_vaccination',
                              'Feel_it_is_important_that_I_get_vaccinated',
                              'Open_to_the_idea_of_receiving_the_flu_jab,_in_principle']]

    vacc_trust['combined_index']=0.0000
    for col in list(vacc_trust.columns):
        print(col)
        vacc_trust['combined_index']=vacc_trust['combined_index']+vacc_trust[col].astype(float)

    theme_data['vacc_trust'] = vacc_trust[['combined_index']]

    past_exper = labeled_data[['Bad_experience_with_vaccines_or_injections_as_a_child',
                              'Bad_experience_with_vaccines_or_injections_as_an_adult']]

    past_exper['combined_index']=0.0000
    for col in list(past_exper.columns):
        print(col)
        past_exper['combined_index']=past_exper['combined_index']+past_exper[col].astype(float)

    theme_data['past_exper'] = past_exper[['combined_index']]

    len(theme_data.index)

    for col in list(theme_data.columns):
        theme_data[col]=np.where(theme_data[col]<=theme_data[col].describe()['50%'],0,1)

    theme_data['Adherence'] = np.where(data['Adherence_Score']<=40,0,1)

    theme_data_orignal= theme_data.copy()

    for col_name in list(theme_data_orignal.columns)[0:len(theme_data_orignal.columns)-1]:
        for col_val in [0,1]:
            for adh in [0,1]:
                theme_data = theme_data_orignal.copy()
                if adh == 1:
                    Adherent_str = "Adherent"
                    Adherent_flag = 1
                else:
                    Adherent_str = "Non Adherent"
                    Adherent_flag = 0
                theme_data=theme_data[theme_data[col_name]==col_val]
                theme_data = theme_data.drop([col_name],axis=1)
                total_respondent_count = len(theme_data.index)
                data_x =  theme_data.drop(['Adherence'],axis=1)
                data_y = theme_data['Adherence']
                from sklearn import cross_validation
                from sklearn import metrics
                X_train, X_test, y_train, y_test = cross_validation.train_test_split(data_x, data_y, test_size=0.4, random_state=42)

                from sklearn import tree
                clf = tree.DecisionTreeClassifier()

                #from sklearn.multiclass import OneVsRestClassifier
                #from sklearn.svm import SVC
                #clf = OneVsRestClassifier(SVC(kernel="linear"))
                #from sklearn.ensemble import RandomForestClassifier
                #clf = OneVsRestClassifier(RandomForestClassifier(n_estimators=10))
                #from sklearn.naive_bayes import MultinomialNB
                #clf = OneVsRestClassifier(MultinomialNB(alpha=0.01))
                #from sklearn import svm
                #clf = svm.SVC(kernel="linear")


                #from sklearn.ensemble import RandomForestClassifier
                #clf = RandomForestClassifier(n_estimators=10)
                clf.fit(X_train, y_train) 
                y_pred = clf.predict(X_test)

                print("Overall Accuracy = %s" %metrics.accuracy_score(y_pred,y_test))
                print("Classification Matrix")
                print(metrics.classification_report(y_pred,y_test))



                n_nodes = clf.tree_.node_count
                children_left = clf.tree_.children_left
                children_right = clf.tree_.children_right
                feature = clf.tree_.feature
                threshold = clf.tree_.threshold

                node_depth = np.zeros(shape=n_nodes, dtype=np.int64)
                is_leaves = np.zeros(shape=n_nodes, dtype=bool)
                stack = [(0, -1)]  # seed is the root node id and its parent depth
                while len(stack) > 0:
                    node_id, parent_depth = stack.pop()
                    node_depth[node_id] = parent_depth + 1

                    # If we have a test node
                    if (children_left[node_id] != children_right[node_id]):
                        stack.append((children_left[node_id], parent_depth + 1))
                        stack.append((children_right[node_id], parent_depth + 1))
                    else:
                        is_leaves[node_id] = True

                node_indicator = clf.decision_path(data_x)
                leave_id = clf.apply(data_x)

                data_node = pd.DataFrame(is_leaves)

                data_node.columns =['Is_Leaf_Node']

                data_node['Sample']=clf.tree_.n_node_samples

                data_node=data_node[data_node['Is_Leaf_Node']==True]

                data_node['Node'] = data_node.index

                data_node=data_node.sort_values(by=['Sample'],ascending=False)

                data_node.index = range(0,len(data_node.index))

                def adh_nadh(x):
                    inter=clf.tree_.value[x]
                    if inter[0][0]>inter[0][1]:
                        return "Non Adherent"
                    else:
                        return "Adherent"
                try:
                    data_node['Sample_Type']=data_node['Node'].apply(adh_nadh)
                
                except :
                    continue

                data_node.columns=['Is_Leaf_Node', 'Sample', 'Leaf_Node', 'Sample_Type']

                col_list = list(data_x.columns)

                for leaf_node in data_node['Leaf_Node'].tolist():
                    node_trav = leaf_node
                    node_id_list=[]
                    node_feature_name=[]
                    threshold_val_list=[]
                    threshold_sign_list=[]
                    respondent_count_list = []

                    while (node_trav!=0):
                        try:
                            itemindex = np.where(children_left==node_trav)
                            index =itemindex[0][0]
                            node_id_list.append(itemindex[0][0])
                            node_feature_name.append(col_list[feature[itemindex[0][0]]])
                            threshold_val_list.append(threshold[itemindex[0][0]])
                            threshold_sign_list.append("<=")
                            respondent_count_list.append(clf.tree_.value[index][0][0])
                        except:
                            itemindex = np.where(children_right==node_trav)
                            index =itemindex[0][0]
                            node_id_list.append(itemindex[0][0])
                            node_feature_name.append(col_list[feature[itemindex[0][0]]])
                            threshold_val_list.append(threshold[itemindex[0][0]])
                            threshold_sign_list.append(">")  
                            respondent_count_list.append(clf.tree_.value[index][0][1])
                        node_trav = itemindex[0][0]

                    path_data=pd.DataFrame([node_id_list,node_feature_name,threshold_val_list,threshold_sign_list,respondent_count_list]).T

                    path_data['order'] = list(reversed(range(0,len(path_data.index))))

                    path_data['Leaf_Node'] = leaf_node

                    path_data.columns=['Node_id','Node_feature_name','threshold_val','threshold_sign','respondent_count','order','Leaf_Node']

                    try:
                        master_path_data = pd.concat([master_path_data,path_data])
                    except:
                        master_path_data = path_data.copy()



                data_node= data_node[data_node['Sample_Type']==Adherent_str]

                data_node.index = range(0,len(data_node.index))

                master_path_data=master_path_data[master_path_data['Leaf_Node']==data_node['Leaf_Node'][0]].sort_values(by='order')

                master_path_data.index=range(0,len(master_path_data.index))

                theme_data = theme_data[theme_data['Adherence']==Adherent_flag]
                respondent_count_li =[]
                i=0
                for col in master_path_data['Node_feature_name'].tolist():
                    str1='theme_data = theme_data[theme_data['+"'"+col+"'"+']'+str(master_path_data[master_path_data['Node_feature_name']==col]['threshold_sign'][i])+str(master_path_data[master_path_data['Node_feature_name']==col]['threshold_val'][i])+"]"   
                    exec(str1)
                    respondent_count_li.append(len(theme_data.index))
                    i=i+1

                master_path_data['respondent_count']=respondent_count_li

                master_path_data['Filter']=col_name
                master_path_data['Value']=col_val
                master_path_data['Adh']=Adherent_str
                master_path_data['total_respondent_count']=total_respondent_count

                try:
                    master_analysis_data = pd.concat([master_analysis_data,master_path_data])
                except:
                    master_analysis_data=master_path_data



    master_analysis_data=master_analysis_data[['Filter','Value','Adh','order','Node_feature_name','respondent_count','threshold_sign','total_respondent_count']]

    index_col_map=pd.DataFrame(['cues_to_action','vacc_trust','doctor_rel', 'knowledge','attitude','Anticipated_Regret', 'risk_perc','past_exper','trust_info_chan','barriers','Availablity']   )

    index_col_map['neg_map']= ['No Cues for Action' , ' No Trust on Vaccination','Bad Doctor Relationship','Less Knowledge','Bad Attitude','Low Anticipated Regret','Low risk perception','Good past Experience','Low trust on Info Channel','Less Barriers','Low Availablity']

    index_col_map['pos_map']= ['Given Cues for Action' , ' Have Trust on Vaccination','Good  Doctor Relationship','Good Knowledge','Good Attitude','High Anticipated Regret','High risk perception','Bad past Experience','High trust on Info Channel','More Barriers','High Availablity']

    index_col_map.columns=['Node_feature_name', 'neg_map', 'pos_map']

    master_analysis_data=pd.merge(master_analysis_data,index_col_map,on='Node_feature_name',how='left')

    master_analysis_data['Col_Name']= np.where (master_analysis_data['threshold_sign']=="<=",master_analysis_data['neg_map'],master_analysis_data['pos_map'])

    master_analysis_data['Adh']=np.where(master_analysis_data['Adh']=="Adherent","Adhere","Donot Adhere")
    master_analysis_data['Country'] = con

    try:
        country_master_analysis_data = pd.concat([country_master_analysis_data,master_analysis_data])
    except:
        country_master_analysis_data=master_analysis_data
```
